# **Importing the bs4 library**

```python
from bs4 import BeautifulSoup
```

```python
with open("links.html") as file:
    data = file.read()
```

# Making the object

```python
soup = BeautifulSoup(data, "html.parser")
```

# Scraping the whole page

```python
print(soup.prettify())
```

# Displaying the title of page

```python
print(soup.title)
print(soup.title.string)
```

# Displaying all the anchor tags

```python
anchor_tags = soup.find_all(name="a")
for _ in anchor_tags:
    print(_)
for _ in anchor_tags:
    print(_.string)
```

# Displaying all the ul and li 

```python
ul_item = soup.find_all(name="ul")
for _ in ul_item:
    print(_)
for _ in ul_item:
    print(_.get_text())
```

# Displaying the paragraphs tags and text

```python
paragraph = soup.select(selector=".para")
print(paragraph)
```

```python
for _ in paragraph:
    print(_.string)
```

# **Scraping a live website**

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://ntu.edu.pk/"
response = requests.get(url=url)
data = response.text
soup = BeautifulSoup(data,"html.parser")
print(soup.title)
print(soup.title.string)
```

```python
anchor_tags = soup.find_all(name="a")
for _ in anchor_tags:
    print(_)
```

```python
anchor_tags = soup.find_all(name="a",class_="small")
for _ in anchor_tags:
    print(_)
```

```python
anchor_tags = soup.find_all(name="a",class_="h6")
for _ in anchor_tags:
    print(_.text.strip())
```
